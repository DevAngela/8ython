# 모든코인 상승장? 하락장?
> 최고가와 최저가의 차이를 변동폭으로 정의할 때 (시가 + 변동폭)이 최고가 보다 높을 경우 "상승장", 그렇지 않은 경우 "하락장" 문자열을 출력하라.


|Key Name|Description|
|------|---|

|closing_price|최근 24시간 내 마지막 거래금액|
|min_price|최근 24시간 내 최저 거래금액|

'24H_fluctate_rate': +이거나 -이거나 이 값으로 오늘의 장 상황만을 볼 수도 있지만,
우리는 전체 시장 상황을 그냥 둔하게 이렇게 정의해봄

```python
import requests
from pprint import pprint
url = "https://api.bithumb.com/public/ticker/all"
data = requests.get(url).json()['data']

for key, value in data.items():
    #print(type(value)) 마지막 친구가 str이고 나머지는 다 dict 타입이니 이걸 이용해서 우선 전처리한다.
    if type((value)) == type(dict()):
        #다 + pr만, 가격이 소수점으로 된 애들이 있어서 float로 가격값 가져오기
        max_price = float(value['max_price'])
        min_price = float(value['min_price'])
        price_range = max_price - min_price
        opening_price = float(value['opening_price'])
        if opening_price + price_range > max_price:
            print('상승장')
        else:
            print('하락장')
    else:
        #이렇게 원하지 않는 데이터(str타입인 date)는 continue로 가세요~ 날려주면 된다.
        continue
```

```python
import requests
from pprint import pprint
url = "https://api.bithumb.com/public/ticker/all"
data = requests.get(url).json()['data']


def gazua(op, minp, maxp):
    if (op + (maxp - minp)) > maxp:
        return '상승장'
    else:
        return '하락장'

# print(gazua(12803000, 11341000, 11000000, 13000000))

# 시장 전체 데이터
def check_gazua(data):
    # 모든 토큰 이름 : 상승장 or 하락장
    # gazua()
    # result = {'BTC': '하락장', 'ETH': 하락장, ...}
    result = {}
    for key, value in data.items():
        if type(value) == type(dict()):
            max_price = float(value['max_price'])
            min_price = float(value['min_price'])
            opening_price = float(value['opening_price'])
            result[key] = gazua(opening_price, min_price, max_price)
    
    return result

print(check_gazua(data))
```

# UBD
> movies는 영화제목이 key로 누적관객수가 value인 딕셔너리이다. 
>
> 자전차왕 엄복동의 누적관객수는 172212명이고 172212명을 1UBD라고 할때 80UBD를 넘지 못하는 영화를 출력하라.

```python
movies = {
    "7번방의선물":12811206,
    "괴물":13019740,
    "국제시장":14257115,
    "극한직업":16261018,
    "도둑들":12983330,
    "명량":17613682,
    "베테랑":13414009,
    "신과함께-죄와벌":14410754,
    "아바타":13624328,
    "어벤져스:엔드게임":13901423,
}

def UBD(movies):
    UBD = 172212
    result = dict()
    
    #1. movies 딕셔너리를 돌면서 value가 80UBD 넘는지 봐야 함
    for key, value in movies.items():
        if value > 80*UBD:
            result[key] = '{:.4} UBD'.format(value/UBD)
    #2. 80UBD 넘는 것은 key를 저장해서 출력하기
    return result

UBD(movies)
```
